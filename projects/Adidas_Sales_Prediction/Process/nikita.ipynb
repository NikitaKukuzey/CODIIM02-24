```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("adidas_sales.csv", sep=";")
df = df.drop(df.columns[7:9], axis=1)
df.head(10)
```

```
names = ['Retailer', 'Region', 'Product', "State", 'City', "Operating Margin", 'Sales Method'] 
for name in names:
        df[name].hist(bins=df[name].nunique() * 2 - 1, label=name)
        plt.legend()
        plt.show()
```

```
names = ['Retailer', 'Region', 'Product', "State", 'City', "Operating Margin", 'Sales Method'] 
##colors = ['cyan', 'magenta', 'lime', 'pink']
##для строки выше надо добавить colors=colors в plt.pie
for name in names:
    city_counts = df[name].value_counts(normalize=True) * 100
    plt.figure(figsize=(8, 8))
    plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%')
    plt.show()
```
