```python
import pandas as pd
import xgboost as xgb
```

```python
d1=pd.read_csv('/transactions_train.csv')
```

```python
d2=pd.read_csv('/train_target.csv')
```

* client_id - уникальный идентификатор клиента
* trans_date - дата совершения транзакции
* small_group - категория покупки
* amount_rur - сумма транзакции

```python
af=d1.groupby('client_id')['amount_rur'].agg(['sum','mean','std','min','max']).reset_index()
af.head()
```

```python
dt=d1.groupby(['client_id','small_group'])['amount_rur'].count()
```

```python
cdt=dt.reset_index().pivot(index='client_id', \
                           columns='small_group',values='amount_rur')
```

```python
cdt=cdt.fillna(0)
```

```python
cdt.columns=['small_group_'+str(i) for i in cdt.columns]
```

```python
cdt.head()
```

```python
train=pd.merge(d2,af,on='client_id')
```

```python
train=pd.merge(train,cdt.reset_index(),on='client_id')
```

```python
train.head()
```

```python
t1=pd.read_csv('/transactions_test.csv')

t2=pd.read_csv('/test.csv')
```

```python
aft=t1.groupby('client_id')['amount_rur'].agg(['sum','mean','std','min','max']).reset_index()
```

```python
aft.head()
```

```python
tdt=t1.groupby(['client_id','small_group'])['amount_rur'].count()
```

```python
ttt=tdt.reset_index().pivot(index='client_id', columns='small_group',values='amount_rur')
```

```python
ttt=ttt.fillna(0)
```

```python
ttt.columns=['small_group_'+str(i) for i in ttt.columns]
```

```python
ttt.head()
```

```python
test=pd.merge(t2,aft,on='client_id')
```

```python
test=pd.merge(test, ttt.reset_index(),on='client_id')
```

```python
common_features=list(set(train.columns).intersection(set(test.columns)))
```

```python
y_train=train['bins']
X_train=train[common_features]
X_test=test[common_features]
```

```python
X_wo = X_train
y_wo = y_train
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import mean_squared_error,r2_score

import warnings
```

```python

X_wo.columns = X_wo.columns.astype(str)
```

```python
lbl = LabelEncoder()
lbl.fit(y_wo)

y_enc = lbl.transform(y_wo)


### Разбираемся с признаками
scl = StandardScaler()
scl.fit(X_wo)
X_scaled = scl.transform(X_wo)

X_train_wo, X_test_wo, y_train_wo, y_test_wo = train_test_split(X_wo, y_enc, test_size=0.2, random_state=42)

### Прописываем параметры для нашей модели
params = {
          'min_samples_split': 10,
          'n_estimators': 100,
          'max_depth':100,
          'learning_rate': 0.05,
          'min_samples_leaf':16,
          'min_samples_split':16
          }
```

```python
### Тренируем
gbс = GradientBoostingClassifier(**params)
gbс.fit(X_train_wo,y_train_wo)
```

```python
### Вычисляем точность
train_accuracy_score=gbс.score(X_train_wo,y_train_wo)
print(train_accuracy_score)

test_accuracy_score=gbс.score(X_test_wo,y_test_wo)
print(test_accuracy_score)

### Предсказание
y_pred_wo = gbс.predict(X_test_wo)
```
